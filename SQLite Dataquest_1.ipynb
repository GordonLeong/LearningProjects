```python
%%sql
SELECT * FROM facts LIMIT 5
```

```python
%%sql
SELECT MIN(population) min_pop, MAX(population) max_pop, 
MIN(population_growth) min_popgrowth, MAX(population_growth) max_popgrowth
FROM facts
```

```python
%%sql
SELECT name 
FROM facts 
WHERE population = (SELECT MIN(population) FROM facts)
```

```python
%%sql
SELECT name 
FROM facts 
WHERE population = (SELECT MAX(population) FROM facts)
```

```python
%%sql 
SELECT AVG(population), AVG(area) FROM facts
```

```python
%%sql
SELECT name 
FROM facts
WHERE population > (SELECT AVG(population) FROM facts) 
AND area < (SELECT AVG(area) FROM facts)
```

```python
%%sql

SELECT name, ROUND(CAST(area_water as float)/CAST(area_land as float),4) as water_to_land
from facts
ORDER BY water_to_land DESC
```

```python
%%sql

SELECT name, (population_growth*population) as abs_growth
from facts
order by abs_growth desc
```

```python
%%sql

SELECT name FROM facts
WHERE death_rate > birth_rate
```
